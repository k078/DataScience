```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
pen = sns.load_dataset('penguins')
```

```python
pen.head()
```

```python
sns.pairplot(pen)
```

```python
pen = pen.dropna()
```

```python
features = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
```

```python
km = KMeans(n_clusters=3, random_state=43).fit(pen[features])
```

```python
pen['cluster'] = km.predict(pen[features])
```

```python
pen.cluster.value_counts()
```

```python
sns.pairplot(pen, hue='cluster')
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
```

```python
metrics.silhouette_score(pen[features], km.labels_, metric='euclidean')
```

```python
sns.pairplot(pen, hue='species')
plt.show()
```

```python
contingency_table = pen.groupby(['species', 'cluster']).size().unstack('cluster', fill_value=0)
```
